```python
import numpy as np
zDA = 30
iDA = 84
jDA = 96
delta_z = np.zeros((iDA,jDA,zDA))
```

```python
def func_square_delta_z(delta_z):
    sum_delta_z = 0
    for z in range(zDA):
        for i in range(iDA):
            for j in range(jDA):
                sum_delta_z = sum_delta_z + delta_z[i][j][z]**2
    return sum_delta_z
```

```python
func_square_delta_z(delta_z)
```
